### Real-world phenomenon chosen

The real-world phenomenon I have chosen is the attitude of primary school children to Santa Claus. I decided the most interesting variable related to this is the age of children when they stop writing letters to Santa Claus (Letter). 

On investigation of the problem I reasoned that the Sex of the child (sex); whether or not the child has older siblings (sibs); and the number of hours per week the child has access to a mobile device (hours) all have a relationship to the age they stopped writing letters to Santa Claus.

The Sex (sex) variable will be M/F and we can use the binomial distribution; The Siblings (sibs) variable will be Yes/No and here again we can use the binomial distribution; The Hours per week (hours) variable will be non-negative real number with two decimal places.

After extensive research at home observing my children and interrogating other parents, I find that primary school students access mobile devices on average five hours per week with a standard deviation of one hour and that a normal distribution is an acceptable model for such a variable.

*Hypotheses:*
An Hypothesis can be defined as a tentative explanation for an observation, phenomenon, or scientific problem that can be tested by further investigation. This means a hypothesis is the stepping stone to a soon-to-be proven theory. [1]
In this analysis we will have a nuber of different hypotheses to test using various independent variables (cause) and dependent variables (effect).

- The first hypothesis is a simple hypothesis prediction of the relationship between two variables; whether there is relationship between the age of the child, their sex, older siblings or not AND if they wrote a letter to Santa Claus.
- The second hypothesis is a logical hypothesis prediction with a proposed explanation possessing limited evidence i.e. that girls under between the age of 7 and 12 are more likely to write a letter to Santa Claus than boys.
- The third hypothesis is a null hypothesis that there is no relationship between the variables

### Synthesise a data set 

Here I will create my dataframe using the Pandas library and utilising the variables we have discussed above.


```python
# Create a dataframe with all of the variables  and their data randomly generated

import pandas as pd
df=pd.DataFrame(data={'Age': age, 'Hours on mobile device': hours, 'Sex': sex, 'Older Siblings': sibs})

df.head(10)
```

```python
# WIP
# Replacing numerical values with strings [4]:

df ['Sex'].replace({0:'Male', 1:'Female'}, inplace=True)
df ['Older Siblings'].replace({0:'Yes', 1:'No'}, inplace=True)
df.head(10)
```

```python
# change order of the columns

df = df[['Sex','Older Siblings','Age','Hours on mobile device']]
df.head()
```


#### Analysis: 
Below I have added an additional column, the sum of age and average weekly hours spent on a mobile device. From researching online and based on the average hours per week spent online of 5 hours, I have concluded that if the value of Column **Age+Hours** is greater (>) than 14 hours than the liklihood is that the child did not write a letter to Santa Claus.

```python
# adding additional column based on the other columns [7]:
df['Age+hours'] = (df.Age + df['Hours on mobile device'])
df.head(20)
```

```python
# adding additional column based on the other columns [6]:
# where True = letter was written
# and False = No letter was written

import numpy as np
df['Letter'] = np.where(df['Age+hours']<14.00, True, False)
df.head(30)
```

#### Evaluate:

I am going to use the K nearest neighbors algorithim to evaulate the dataset, specifically the numeric variables 'Age' and 'Hours spent on a mobile device' to see if we can predict the result in 'Actual_Letter' using data similar to but not from my actual dataset.

```python
# Using KNN to evaluate to dataset
import sklearn.neighbors as nei
import pandas as pd
import sklearn.model_selection as mod
```

```python
# Here we are training our KNN algorithim with a set of inputs to which we know the outcome (from my synthesised dataframe above)
# Inputs and Outputs are subsets of the dataframe
# KNN algorithim will build up a model based on our existing values which should allow us to make a prediction with reasonable accuracy

inputs = df[[ 'Age', 'Hours on mobile device']]
outputs = df['Letter']
```

```python
# We use the KNeighborsClassifier function to create an object which returns the (5 in this case as K=5) closest datapoints 
# to a selected datapoint
knn = nei.KNeighborsClassifier(n_neighbors=5)
```

```python
# This trains the algorithim on my new subset isng the knn.fit method
knn.fit(inputs, outputs)
```

```python
# Check accuracy using the row at index 12
# # The sum of Age' & 'Hours on mobile device is 13.26 so we would expect value returned as True

df.loc[12]
```

```python
# enter values for Age and Hour to predict the response
# The sum of Age' & 'Hours on mobile device is 14.68 so we would expect value returned as False
# And thankfully Knn.predicts correctly

knn.predict([[9, 5.68]])
```

```python
# enter values for Age and Hour to predict the response
# The sum of Age' & 'Hours on mobile device is 12,97 so we would expect value returned as True
# And again thankfully Knn.predicts correctly

knn.predict([[8, 4.97]])
```

```python
# Evaluate KNN
# Here we are comparing the knn predicts function result with the actual output
# It shows that knn predicts got it right 95 times (out of 100), and there for wrong 5 times out of the 100
# result below sum of True responses was 95/100
# The reason for this if we look at the pair plot is that there are a number of values for 'Age+Hours' close to 14
# knn predicts the 5 nearest neighbours but these may be on the wrong side of 14 
# e.g. one data point is at index #28 with a value of 13.85, and its nearest neighbours might be predominately > 14 and so
# knn.predict will return a False response
(knn.predict(inputs) == outputs).sum()
```

```python
# here we are using the sklearn.model_selection we imported above as mod to split the training set into a testing subset
# for comparison purposes.
# we split the training and test set 50/50
# then we re-create the classifier
# fit the new Classifier thsi time using inputs_train AND outputs_train
# we use knn.predict again to compare inputs_test (which the classifier has not seen before) WITH outputs_test
# we are giving the classifier 50 inputs it hasnt seen before and asking it to predict the response
# In this case it returns 48/50 correct which is a decent result, but if we re-run we can get a different result
# because the train test split has a random element to it. [9]


inputs_train, inputs_test, outputs_train, outputs_test = mod.train_test_split(inputs, outputs, test_size=0.50)
```

```python
knn = nei.KNeighborsClassifier(n_neighbors=5)
knn.fit(inputs_train, outputs_train)
```

```python
(knn.predict(inputs_test) == outputs_test).sum()
```
