The real-world phenomenon I have chosen is the attitude of primary school children to Santa Claus. I decided the most interesting variable related to this is the age of children when they stop writing letters to Santa Claus (Actual_Letter). 

On investigation of the problem I reasoned that the Sex of the child (sex); whether or not the child has older siblings (sibs); and the number of hours per week the child has access to a mobile device (hours) all have a relationship to the age they stopped writing letters to Santa Claus.

As metioned above I found that primary school students access mobile devices on average five hours per week with a standard deviation of one hour and that a normal distribution is an acceptable model for such a variable

```python
# generates a set of hours of mobile device usage with mean of 5 using the normal distribution
hours = np.random.normal(5, 1, 100)
hours
```

Sex of the child....


```python
# generates an array of the sex of the child using the uniform distribution
# we will convert the integers to strings later
# Male = 0 
# Female = 1

sex = np.random.binomial(1, .5, 100)
sex
```

Whether a child has older siblings or not....


```python
# generates an array Yes/No answers on whether a child has older siblings using the uniform distribution
# we will convert the integers to strings later
# Yes = 0 
# No = 1

sibs = np.random.binomial(1, .5, 100)
sibs
```

### Synthesise a data set 

Here I will create my dataframe....


```python
# Create a dataframe with all of the variables  and their data randomly generated

import pandas as pd
df=pd.DataFrame(data={'Age': age, 'Hours on mobile device': hours, 'Sex': sex, 'Older Siblings': sibs})

df.head(10)
```

```python
# WIP
# Replacing numerical values with strings [4]:

df ['Sex'].replace({0:'Male', 1:'Female'}, inplace=True)
df ['Older Siblings'].replace({0:'Yes', 1:'No'}, inplace=True)
df.head()
```

```python
# change order of the columns

df = df[['Sex','Older Siblings','Age','Hours on mobile device']]
df.head()
```


Below I have added an additional column, the sum of age and average weekly hours spent on a mobile device. From researching online and based on the average hours per week spent online of 5 hours, I have concluded that if the value of Column **Age+Hours** is > than 14 than the liklihood is that the child did not write a letter to Santa Claus.

```python
# adding additional column based on the other columns [5]:
df['Age+hours'] = (df.Age + df['Hours on mobile device'])
df.head(10)
```

```python
# generates is a list of 100 Y/N values drawn from the allowed values - each with equal chance of being drawn
# weighting???

wrote_letter = ['Yes', 'No']

letter = [np.random.choice(wrote_letter) for i in range(100)]
letter

```

```python
# added new column to the dataframe, I need to somehow correleate this new column 'Wrote Letter' with 'Age+hours'


df['Wrote letter'] = (letter)
df.head(10)
```

```python
# WIP
# adding additional column based on the other columns [6]:
# where True = letter was written
# and False = No letter was written

import numpy as np
df['Actual_Letter'] = np.where(df['Age+hours']<14.00, True, False)
df
```

```python
# discard the erroneous 'Wrote letter' column which was just a randonmly generated arrary
# and not based on any values from the other columns
# unlike 'Actual Letter' which is based on the value in 'Age+hours'
df.drop(['Wrote letter'], axis = 1, inplace = True)
df
```

```python
# Here we are selecting only rows that have Actual Letter = True, or children that have sent letters to Santa
df.loc[df['Actual_Letter'] == True]
```

```python
# get an actual count of letters written/not written
# where True = letter written & False = No letter written

df['Actual_Letter'].value_counts()
```
